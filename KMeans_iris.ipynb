```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as py
import plotly.graph_objs as go
from sklearn.cluster import KMeans
```

```python
import warnings
warnings.filterwarnings('ignore')

```

```python
data=pd.read_csv('/content/Iris_1.csv')
data
```

```python
data.isnull().sum()
```

```python
x=data.iloc[:,[2,4]].values
x
```

```python
w=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=0)
  kmeans.fit(x)
  w.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),w)
plt.title('Elbow Method')
plt.xlabel('no.of.clusters')
plt.ylabel('w')
plt.show()
```

```python
kmeansmodel=KMeans(n_clusters=3,init='k-means++',random_state=0)
y_kmeans=kmeansmodel.fit_predict(x)
```

```python
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 50,c = 'red', label = 'Cluster 1',edgecolor='black')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 50 ,c = 'blue', label = 'Cluster 2',edgecolor='black')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 50, c = 'green', label = 'Cluster 3',edgecolor='black')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],marker='*', s = 200, c = 'black', label = 'Centroids',edgecolor='black')
plt.title('petalwidth')
plt.xlabel('petallength')
plt.ylabel('species')
plt.legend()
plt.show()
```
