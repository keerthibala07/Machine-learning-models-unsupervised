```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Mall_Customers.csv')
```

```python
features=['Annual Income (k$)','Spending Score (1-100)']
x=df[features]
x
```

```python
x_std=StandardScaler().fit_transform(x)
```

```python
dbscan=DBSCAN(eps=0.5,min_samples=5)
clusters=dbscan.fit_predict(x_std)
clusters
```

```python
centroids=pd.DataFrame(x_std).groupby(clusters).mean().values
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'],c=clusters,cmap='viridis',edgecolor='black')
plt.title("DBSCAN")
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```
