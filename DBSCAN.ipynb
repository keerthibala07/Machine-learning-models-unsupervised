```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('/content/Iris_1.csv')
df
```

```python
features=['PetalLengthCm',	'PetalWidthCm']
x=df[features]
x
```

```python
x_std=StandardScaler().fit_transform(x)
```

```python
dbscan=DBSCAN(eps=1,min_samples=3)
clusters=dbscan.fit_predict(x_std)
clusters
```

```python
centroids=pd.DataFrame(x_std).groupby(clusters).mean().values
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x['PetalLengthCm'],x['PetalWidthCm'],c=clusters,cmap='viridis',edgecolor='black')
plt.title("DBSCAN")
plt.xlabel('PetalLengthCm')
plt.ylabel('PetalWidthCm')
plt.show()
```
